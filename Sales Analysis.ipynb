```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

## Data importing from csv to python.

## Combinig the 12 Months data.

```python
files = [file for file in os.listdir('TwelveMonthsSalesData')]

all_df=pd.DataFrame()

for file in files:
    df = pd.read_csv('./TwelveMonthsSalesData/'+file)
    all_df = pd.concat([all_df,df])
all_df    
```

## Making the month column

```python
all_df['Month'] = all_df['Order Date'].str[0:2]
```

```python
all_df
```

## Cleaning of data

```python
all_df = all_df.dropna()
all_df
```

```python
temp_df = all_df[all_df['Quantity Ordered'].str[0:17] == 'Quantity Ordered']
temp_df
```

```python
all_df = all_df[all_df['Quantity Ordered'].str[0:17] != 'Quantity Ordered']
all_df
```

## Changing of data type.

```python
all_df['Quantity Ordered'] = all_df.loc[:,'Quantity Ordered'].astype(int)
```

```python
all_df['Price Each'] = all_df.loc[:,'Price Each'].astype(float)
```

```python
all_df.info()
```

## Making a Sales column.

```python
all_df['Sales'] = all_df['Quantity Ordered'] * all_df['Price Each']
```

```python
all_df
```

## adding as city column

```python
# add a city column
all_df['City'] = all_df['Purchase Address'].apply(lambda x:x.split(',')[1]+','+x.split(',')[2])
```

```python
all_df
```

```python
result1 = all_df.groupby(all_df['City']).sum()
result1
```

```python
all_df
```

```python
sum_of_sales = all_df.groupby(['Month']).sum()
sum_of_sales
```

```python
results = sum_of_sales['Sales']
results
```

## In which month sales is maximum?

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.figure(figsize=(5,3),dpi=300)
plt.title('Sales report')

plt.xlabel('Months')
plt.ylabel('Sales in USD ($) in millions')

plt.bar(months,results,color='green')


plt.show()
```

## Changing the order date column into date and time format.


```python
all_df['Order Date'] = pd.to_datetime(all_df['Order Date'])
all_df['Hour'] = all_df['Order Date'].dt.hour
#all_df['Minute'] = all_df['Order Date'].dt.minute
```

```python
all_df
```

```python
order_time = all_df.groupby(by='Hour').count()
```

## At which time the ordered calls are maximum?

```python
hour = [hour for hour, df in all_df.groupby('Hour')]

plt.figure(figsize=(6.5,2.5),dpi=160)

plt.title('TIme vs orders called')

plt.plot(hour,order_time['Order ID'],label = 'Ordered Calls')

plt.xticks(hour)

plt.xlabel('Hours')
plt.ylabel('No. of Orders')
plt.grid()
plt.legend()
plt.show()
```
